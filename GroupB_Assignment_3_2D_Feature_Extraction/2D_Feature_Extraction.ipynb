```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load an image
image = cv2.imread('input_image.jpg')
if image is not None:
# Convert the image to grayscale for texture analysis
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
# Feature extraction
color_features = image.reshape((-1, 3)) # Reshape the color image into a 2D array

texture_features = gray_image.reshape(-1) # Reshape the grayscale image into a 1D array
```

```python
# Plot color histogram
plt.figure(figsize=(12, 6))
plt.subplot(121)
plt.hist(color_features, bins=256, color=['blue', 'green', 'red'])
plt.title('Color Histogram')
plt.xlabel('Pixel Value')
plt.ylabel('Frequency')
```

```python
if image is not None:
    # Plot texture histogram
    plt.subplot(122)
    plt.hist(texture_features, bins=256, color='gray')
    plt.title('Texture Histogram')
    plt.xlabel('Pixel Value')
    plt.ylabel('Frequency')

    plt.show()
else:
    print('Image could not be found or could not be loaded.')
```
